## 1b: Swarm Plot 

**INSTRUCTION: Demonstrate the age in both genders separeted by manner of death by a swarm plot.**

```python
# with seaborn
plt.figure(figsize=(10,6))

sns.swarmplot(x="gender", y="age", hue="manner_of_death", data=kill, order=["M", "F"], size=5);
```

## 2: Plot 12 most dangerous cities

```python
city_freq = kill.city.value_counts()[:12]
city_freq 
```

```python
fig, ax =plt.subplots(figsize=(8,6))

ax = sns.barplot(x=city_freq.index, y=city_freq.values)
ax.set_title("Most Dangerous Cities", color="b", fontsize=14)
ax.set_xticklabels(city_freq.index, rotation = 45);


#sns.barplot(x=city_freq.index, y=city_freq.values, data=kill)
#plt.xticks(rotation=45)
#plt.title("Most Dangerous Cities", color ="b");
```

```python
# with seaborn countplot
fig, ax =plt.subplots(figsize=(8,5))
sns.countplot(x="city", data=kill, order=city_freq.index)
ax.set_xticklabels(city_freq.index, rotation = 45)
ax.set_title("Most Dangerous Cities", color="b",fontsize=14);


#sns.countplot(order=city_freq.index, x="city", data=kill)
#plt.xticks(rotation=45)
#plt.title("Most Dangerous Cities", color ="b");
```

## 3: Plot 20 most dangerous states

```python
state_freq = kill.state.value_counts()[:20]
state_freq
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax = sns.barplot(x=state_freq.index, y=state_freq.values)
ax.set_title("Most Dangerous States", color="b", fontsize=15)
ax.set_xticklabels(state_freq.index, rotation = 45);


#sns.barplot(x=state_freq.index, y=state_freq.values, data=kill)
#plt.xticks(rotation=45)
#plt.title("Most Dangerous States", color ="b");


```

```python
fig, ax =plt.subplots(figsize=(8,5))
sns.countplot(x="state", data=kill, order=state_freq.index)
ax.set_xticklabels(state_freq.index, rotation = 45)
ax.set_title("Most Dangerous States", color="b",fontsize=14);


#sns.countplot(order=state_freq.index, x="state", data=kill)
#plt.xticks(rotation=45)
#plt.title("Most Dangerous States", color ="b");
```
